```python
# compute covariance matrix and its corresponding eigenvaluexs/eigenvectors
S = 1/(len(yalefaces)) * (yalefaces @ yalefaces.T)
w, v = np.linalg.eig(S)
```

```python
plt.plot(w, '-o')
plt.show()
```

```python
total_variation = np.trace(S)

percentage = np.ones(len(S))
for i in range(1, len(S)):
    percentage[i] = np.trace(S[:-i, :-i]) / total_variation
```

```python
plt.plot(percentage)
plt.hlines(y=0.99, xmin=0, xmax=len(S), linewidth=2, color='g')
plt.hlines(y=0.95, xmin=0, xmax=len(S), linewidth=2, color='r')
plt.legend(('Percentage', 'Percentage=99%', 'Percentage=95%'))
plt.show()
```

```python
# number of principle components to reprecent 99%
print([percentage[21], percentage[22], percentage[23]])
print('number of principle components to reprecent 99%: ', 22)
```

```python
# number of principle components to reprecent 95%
print([percentage[114], percentage[115], percentage[116]])
print('number of principle components to reprecent 95%: ', 115)
```
