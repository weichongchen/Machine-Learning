```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
x = np.random.uniform(0, 1, n)
e = np.random.randn(n)*0.5
y = x + e
```

```python
# Problem 4.1
plt.scatter(x,y)
plt.show()
```

```python
# Problem 4.2
a = np.linspace(0, 2, 10000)
m = np.zeros(len(a))
for i in range(0,len(a)):
    m[i] = np.linalg.norm(x*a[i] - y)**2
a_min = a[np.where(m==min(m))]
f = np.linalg.norm(x*a_min - y)**2
print(f)
y1 = a_min*x
print(a_min)
plt.scatter(x,y)
plt.plot(x,y1, color='red')
plt.show()
```

```python
x = np.random.uniform(0, 1, n)
e = np.random.randn(n)*0.1
y = 30*((x-0.25)**2)*((x-0.75)**2) + e
```

```python
# Problem 4.4-1
plt.scatter(x,y)
plt.show()
```

```python
d = 4
X = np.zeros((n,d+1))
for i in range(d+1):
    X[:,i] = x**i
Xt = np.transpose(X)
a = np.linalg.inv(Xt @ X) @ Xt @ y
print(a)
```

```python
# Problem 4.4-2
y1 = X @ a
plt.scatter(x,y)
plt.scatter(x,y1, color='red')
plt.show()
```
