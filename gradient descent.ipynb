```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b = [4.5, 6]
x_0 = [0, 0]
x_opt = b

def error_cal(x, x_opt):
    error = np.linalg.norm(x - x_opt) / np.linalg.norm(x_opt)
    return error
```

```python
# Problem 2.2
errors = []
x = np.zeros((2, len(x_opt)))
error = 100
alpha = 1
i = 1
while (error > 0.01) & (i < 15000):
    x[i,:] = x[i-1,:] - alpha * (x[i-1,:] - b) / np.linalg.norm(x[i-1,:]-b)
    error = error_cal(x[i,:], x_opt)
    x_temp = x[i,:]
    x = np.vstack((x, x_temp))
    errors.append(error)
    i += 1

if i==15000:
    print("Does not converge!")
else:
    print(i-1)
plt.plot(range(1,i), errors)
plt.xlabel("Num of Steps")
plt.ylabel("Error")
plt.show()
```
